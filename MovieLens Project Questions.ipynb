```python

pprint.pprint(user_movies_ratings_pd.head())
pprint.pprint(movie_details_pd.head())
pprint.pprint(user_pd.head())
```

```python
user_movies_ratings_pd_sub = user_movies_ratings_pd[['movie id','rating']].groupby(['movie id']).sum()
```

```python
movie_details_pd.columns
```

```python
movie_details_pd_sub= movie_details_pd[['movie id', 'movie title', 'release date', 'unknown', 'Action',
       'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western', 'release year']]

combinedByYear = pd.merge(user_movies_ratings_pd_sub,movie_details_pd_sub,on=['movie id'])


combinedByYear_sub_byGenre = combinedByYear[['release year','rating','unknown', 'Action',
       'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']].groupby(['release year']).sum()
```

```python
combinedByYear_sub_byGenre_counts = combinedByYear[['release year','rating','unknown', 'Action',
       'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']].groupby(['release year']).size().reset_index(name='counts')

combinedByYear_sub_byGenre_counts_final= result = pd.merge(combinedByYear_sub_byGenre, combinedByYear_sub_byGenre_counts, on=['release year'])
```

```python
combinedByYear_sub_byGenre_counts_final

percentageValues_sub = (combinedByYear_sub_byGenre_counts_final[['unknown', 'Action',
       'Adventure', 'Animation', 'Childrens', 'Comedy', 'Crime', 'Documentary',
       'Drama', 'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery',
       'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']].div(combinedByYear_sub_byGenre_counts_final['counts'], axis= 0) * 100)

combinedByYear_sub_byGenre_counts_final_final = pd.concat( [combinedByYear_sub_byGenre_counts_final[['release year']], percentageValues_sub], axis=1) 
```
